```python
import pandas as pd
import pickle
from tqdm import tqdm
```

```python
with open('../../pickle/prot_to_seq_pickle/embedded_vectors_pro_bert_bfd.pickle', 'rb') as handle:
        prot_to_vec = pickle.load(handle)
```

```python
# list_vector = [[prot_to_vec[row.Protein_A].tolist()] for index,row in data.iterrows()]

def return_vec(name):
    return prot_to_vec[name].tolist()
```
