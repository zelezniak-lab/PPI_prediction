```python
from tqdm.auto import tqdm
import torch
from transformers import BertModel, BertTokenizer
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
tokenizer = BertTokenizer.from_pretrained('Rostlab/prot_bert_bfd', do_lower_case=False )
model = BertModel.from_pretrained("Rostlab/prot_bert_bfd")
model.to(device)
```

```python
from tqdm import tqdm
protein_name = []
protein_sequence = []
seq_to_features = {}

for i in tqdm(x):
    
    protein_name.append(i)
    protein_sequence.append(sequence[i])
len( protein_sequence)
```

```python
maxi =0
for i in protein_sequence:
    if len(i)>  maxi:
        maxi  = len(i)
maxi
```

```python
model.to('cpu')
# protein_sequence_1 = [protein_sequence[0:500],protein_sequence[0:500],protein_sequence[500:1000],protein_sequence[1000:1500],protein_sequence[1500:2000],protein_sequence[2000:2500],protein_sequence[2500:]]

ids = tokenizer.batch_encode_plus(protein_sequence[2000:], add_special_tokens=True, pad_to_max_length=True)
input_ids = torch.tensor(ids['input_ids']).to('cpu')
attention_mask = torch.tensor(ids['attention_mask']).to('cpu')
with torch.no_grad():
    embedding = model(input_ids=input_ids,attention_mask=attention_mask)[0]
```

```python


for seq_num in tqdm(range(len(embedding))):
    
    seq_emd = embedding[seq_num][0]
    features.append(seq_emd)
```
